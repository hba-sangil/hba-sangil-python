```python
import random
def rand_prime():
    list = []
    for i in range(0,50+1):
        for j in range(2, i):
            if (i%j) == 0:
                break
        else:
            list.append(i)
    print(random.choice(list))
rand_prime()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
result=tips['day'].value_counts()  #열에 대한 모든 발생횟수(분포)
print(result)
tips.describe()   #수치형 데이터 요약
```

```python
#sex별 tip의 최댓값
male = tips['sex'] == 'male'
female = tips['sex'] == 'female'
max_male = tips['tip'].loc[male].max()
max_female = tips['tip'].loc[female].max()
print(max_male, max_female)
```

```python
result = tips.loc[(tips['tip'] < 2)]
print(result.count())

result1 = tips.loc[tips['size'][0:4]]
result1
```

```python
print(tips['total_bill'].corr(tips['tip']))  #total bill과 tip의 상관계수
tips['tip_percent'] = (tips['tip']/tips['total_bill']).astype('float')
print(tips['tip_percent'][:4])
```

```python
mpg = pd.read_csv("mpg.csv")
print(mpg.iloc[:5])   #위에서부터 5행을 출력
```

```python
mpg['fl'].value_counts().sort_index()  #fl데이터 빈도수를 알파벳 순서대로 출력
```

```python
mpg1 = pd.DataFrame({'fl' : ['c','d','e','p','r'], 'price_fl' : [2.66, 2.23, 2.11, 2.89, 2.49]})
mpg1
```

```python
mpg2 = pd.merge(mpg, mpg1, on='fl')
mpg2
```

```python
mpg2.groupby(['manufacturer', 'class']).mean()
```

```python
mpg2.groupby(['model', 'class']).mean()
```

```python
mpg2.groupby(['manufacturer'])['displ','price_fl'].agg(['count', 'max', 'min', 'mean','std'])  #다중 통계량을 구하는 agg
```

```python
mpg2.groupby(['model'])['displ', 'price_fl'].agg(['count', 'max', 'min', 'mean'])
```

```python
top5_manufacturer=mpg2.groupby(['manufacturer']).sum().sort_values(by=['hwy'],ascending=False).index[:5]
top5_manufacturer
```

```python
top5_model = mpg2.groupby(['model']).sum().sort_values(by=['price_fl'], ascending=False).index[:10]
top5_model
```
