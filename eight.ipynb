# 문제 3

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

titanic에서 pclass, age, sibsp, fare, class 열만 뽑고, 그 중 타이타닉에 탑승한 형제/자매 수(sibsp)가 0인 행들만 추출하여 df에 저장하시오.
이때 기존 index를 0부터 새롭게 출력되도록 하시오.

```python
fd = titanic.loc[titanic.sibsp==0,['pclass','age','sibsp','fare','class']].reset_index(drop=True)
fd
```

### 문제 [3-1] (10점)

```python
df = titanic.loc[:, ["pclass", "age", "sibsp", "fare", "class"]]
df.head(10)
```
